```python
from selenium import webdriver
from selenium.webdriver.common.by import By
import time
driver = webdriver.Chrome(executable_path='C:\\Users\\amini\\Downloads\\chromedriver_win32.zip\\chromedriver.exe')
driver.get('https://www.amazon.in/')
driver.maximize_window()
time.sleep(3)

ES=driver.find_element(By.XPATH,"//a[normalize-space()='Electronics']")
ES.click()

LA=driver.find_element(By.XPATH,"//span[normalize-space()='Laptops & Accessories']")
LA.click()

L=driver.find_element(By.XPATH,"//span[@dir='auto'][normalize-space()='Laptops']")
L.click()

I=driver.find_element(By.XPATH,"//span[@class='a-size-base a-color-base'][normalize-space()='Intel Core i5']")
I.click()

S='document.querySelector(".sg-col-20-of-24.s-matching-dir.sg-col-16-of-20.sg-col.sg-col-8-of-12.sg-col-12-of-16")'

Fn = driver.execute_script('return'+' '+S)

Fn
```

```python
t = Fn.text
```

```python
t
```

```python
raw = t.split('\n')
```

```python
raw
```

```python
#EXTRACTING COLUMNS FROM RAW DATA
```

```python
title = []
for i in raw:
    if len(i)>len('Flat INR 1000 Off on ALL Banks Cards'):
        title.append(i)
```

```python
title
```

```python
rating = []
total_reviews = []
offerprice = []
Actualprice = []
for i in title:
    if i in raw:
        r = raw.index(i)
        rating.append(raw[r+1])
        total_reviews.append(raw[r+2])
        offerprice.append(raw[r+3])
        Actualprice.append(raw[r+4])
```

```python
rating
```

```python
total_reviews
```

```python
offerprice
```

```python
Actualprice
```

```python
import pandas as pd
```

```python
data = pd.DataFrame({'ProductTitle':title,'Rating':rating,'Totalreviews':total_reviews,'OfferPrice':offerprice,'ActualPrice':Actualprice})
```

```python
data
```

```python
data.to_csv("Amazon_data.csv",index=False)
```

```python
# Extract the numerical value from the column using regular expressions
df1['ActualPrice'] = df1['ActualPrice'].str.extract(r'₹([\d,]+)')
```
